# Q1. Write a python program to display all the header tags from wikipedia.org.

```python
wiki=requests.get('https://www.wikipedia.org/')
soup=BeautifulSoup(wiki.content)
head=[]
s=[]
c=1
for i in soup.find_all('a',class_="link-box"):
    head.append(i.text.split()[0])
    s.append(c)
    c+=1
pd.DataFrame({'Wikipedia Headings':head,'S.No':s}).set_index('S.No')
```

# Q2.To display IMDB’s Top rated 100 movies’ data (i.e. name, rating, year of release) and make data frame.

```python
moviepg=requests.get('https://www.imdb.com/chart/top/?ref_=nv_mv_250')
Soup=BeautifulSoup(moviepg.content)

year=[]
rating=[]
title=[]
rank=[]

for i in Soup.find_all('span', class_="secondaryInfo"):
    year.append(i.text)    

for i in Soup.find_all('td' ,class_="ratingColumn imdbRating"):
    rating.append(i.text.replace('\n',''))

for i in Soup.find_all('td', class_="titleColumn"):
    r=i.text.split()[0]
    rank.append(i.text.split()[0])
    
    temp=i.text.replace('\n','')
    title.append(temp.replace(r,''))

    
pd.DataFrame({'Title':title,'Rating':rating,'Year Released':year,'Rank':rank}).set_index('Rank').head(100)
```

# Q7) Write a python program to scrape details of all the posts from coreyms.com. Scrape the heading, date, content and the code for the video from the link for the youtube video from the post.

```python
corey=requests.get('https://coreyms.com/')
soup=BeautifulSoup(corey.content)
heading=[]
date=[]
content=[]
url=[]
c=[]
count=1
for i in soup.find_all('h2',class_="entry-title"):
    heading.append(i.text.replace('\n',''))
    c.append(count)
    count+=1
for i in soup.find_all('time',class_="entry-time"):
    date.append(i.text)
for i in soup.find_all('div',class_="entry-content"):
    content.append(i.text.replace('\n',''))
for i in soup.find_all('iframe', class_="youtube-player"):
    url.append(i['src'])
url.insert(4,None)
pd.DataFrame({'Heading':heading,'Date':date,'Content':content,'Video_url':url,'S.No':c}).set_index('S.No')
```

# Q9) Write a python program to scrape mentioned details from dineout.co.in: 
i) Restaurant name, ii) Cuisine, iii) Location, iv) Ratings, v) Image URL


```python
dine=requests.get('https://www.dineout.co.in/bangalore-restaurants/welcome-back?p=1')
soup=BeautifulSoup(dine.content)
name=[]
loc=[]
food=[]
rating=[]
image=[]
index=[]
count=0

for i in soup.find_all('div',class_="restnt-card restaurant"):
    name.append(soup.find_all('a',class_="restnt-name ellipsis")[count].text)
    loc.append(soup.find_all('div',class_="restnt-loc ellipsis")[count].text)
    food.append(soup.find_all('span',class_="double-line-ellipsis")[count].text)
    image.append(soup.find_all('img',class_="no-img")[count]['data-src'])
    if (count==10):
        rating.append(soup.find('div',class_="restnt-rating rating-5").text)
    elif (count>10):
        rating.append(soup.find_all('div',class_="restnt-rating rating-4")[count-1].text)
    else:
        rating.append(soup.find_all('div',class_="restnt-rating rating-4")[count].text)
    count+=1
    index.append(count)
pd.DataFrame({'Index':index,'Restaurant Name':name,'Location':loc,'Price & Cuisine':food,'Rating':rating,'Image_url':image}).set_index('Index')
```

# Q10) Write a python program to scrape first 10 product details which include product name , price , Image URL from https://www.bewakoof.com/women-tshirts?ga_q=tshirts .

```python
shirt=requests.get('https://www.bewakoof.com/women-tshirts?ga_q=tshirts')
soup=BeautifulSoup(shirt.content)
name=[]
price=[]
image=[]
s=[]
count=0
for i in soup.find_all('div',class_="productCardBox"):
    name.append(soup.find_all('h3', id_="")[count].text)
    price.append(soup.find_all('span', class_="discountedPriceText")[count].text)
    image.append(soup.find_all('img',class_="productImgTag")[count]['src'])
    count+=1
    s.append(count)
pd.DataFrame({'Name':name,'Price':price,'Image_url':image,'S.No':s}).set_index('S.No')
```
