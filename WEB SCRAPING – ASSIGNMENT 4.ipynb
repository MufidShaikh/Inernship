# Q4.Scrape the details of State-wise GDP of India from statisticstime.com. Url = http://statisticstimes.com/ You have to find following details:
A) Rank
B) State
C) GSDP(18-19)
D) GSDP(17-18)
E) Share(2017)
F) GDP($ billion)

```python
driver=webdriver.Chrome(r'C:\Users\shaik\Downloads\chromedriver_win32\chromedriver.exe')
driver.get('https://www.statisticstimes.com/economy/india/indian-states-gdp.php')
time.sleep(3)
```

```python
count=0
rank=[]
c=0
state=[]
gsdp1=[]
gsdp2=[]
share=[]
gdp=[]
for i in driver.find_elements_by_xpath('//td'):
    count+=1
    if count>=4:
        if (count-4)%8==2:
            state.append(i.text)
            c+=1
            rank.append(c)
        if (count-4)%8==3:
            gsdp2.append(i.text)
        if (count-4)%8==4:
            gsdp1.append(i.text)
        if (count-4)%8==5:
            share.append(i.text)
        if (count-4)%8==6:
            gdp.append(i.text)
    if count==8*33+4:
        break
pd.DataFrame({'Rank':rank,'State Name':state,'GSDP(19-20)':gsdp2,'GSDP(18-19)':gsdp1,'Share':share,'GDP':gdp}).set_index('Rank')
```

```python
driver.close()
```

# Q9. Scrape the details most watched tv series of all time from imdb.com. Url = https://www.imdb.com/list/ls095964455/ .You have to find the following details:
A) Name
B) Year span
C) Genre
D) Run time
E) Ratings
F) Votes

```python
driver=webdriver.Chrome(r'C:\Users\shaik\Downloads\chromedriver_win32\chromedriver.exe')
driver.get('https://www.imdb.com/list/ls095964455/')
```

```python
data=[]
year=[]
rt=[]
genre=[]
ratings=[]
votes=[]
for i in driver.find_elements_by_xpath('//h3[@class="lister-item-header"]'):
    temp=i.text
    data.append(temp.split('(')[0])
for i in driver.find_elements_by_xpath('//span[@class="lister-item-year text-muted unbold"]'):
    year.append(i.text)
for i in driver.find_elements_by_xpath('//span[@class="runtime"]'):
    rt.append(i.text)
for i in driver.find_elements_by_xpath('//span[@class="genre"]'):
    genre.append(i.text)
for i in driver.find_elements_by_xpath('//span[@class="ipl-rating-star__rating"]'):
    temp=i.text.replace('Rate','')
    ratings.append(temp)
ratings = list(filter(None, ratings))
count=0
for i in driver.find_elements_by_xpath('//p[@class="text-muted text-small"]'):
    count+=1
    if count%3==0:
        votes.append(i.text)
pd.DataFrame({'Name':data,'Year Span':year,'Runtime':rt,'Genre':genre,'Ratings':ratings,'Votes':votes})
```

```python
driver.close()
```
