# Q10: Write a python program to scrape the salary data for Data Scientist designation. You have to scrape Company name, Number of salaries, Average salary, Minsalary, Max Salary.

```python
driver=webdriver.Chrome(r'C:\Users\shaik\Downloads\chromedriver_win32\chromedriver.exe')
url='https://www.ambitionbox.com/'
driver.get(url)
driver.find_element_by_xpath('/html/body/div[1]/nav/nav/a[4]').click()
driver.find_element_by_xpath('/html/body/div/div/div/main/section[1]/div[2]/div[1]/span/input').send_keys('Data Scientist')
```

```python
driver.find_element_by_xpath('/html/body/div/div/div/main/section[1]/div[2]/div[1]/span/input').click()
driver.find_element_by_xpath('/html/body/div/div/div/main/section[1]/div[2]/div[1]/span/div/div/div[1]').click()
```

```python
exp=[]
company=[]
totals=[]
avgs=[]
mins=[]
maxs=[]
for i in range(0,10,1):
    temp=driver.find_elements_by_xpath('//div[@class="name"]')[i].text.split('\n')
    company.append(temp[0])
    totals.append(temp[1].replace('based on ','').replace(' salaries',''))
    exp.append(driver.find_elements_by_xpath('//div[@class="salaries sbold-list-header"]')[i].text.split('\n')[2].replace(' exp',''))
    avgs.append(driver.find_elements_by_xpath('//p[@class="averageCtc"]')[i].text)
    temp=driver.find_elements_by_xpath('//div[@class="salary-values"]')[i].text.split('\n')
    mins.append(temp[0])
    maxs.append(temp[1])
    
pd.DataFrame({'Company':company,'Total Salary Records':totals,'Experience':exp,'Average Salary':avgs,'Min Salary':mins,'Max Salary':maxs})
```
